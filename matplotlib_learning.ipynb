## Visualiser l'incertitude
### Donnes Discrte

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy + np.random.randn(50)
```

```python
# plot the error
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='silver')
plt.tight_layout()
```

```python
plt.errorbar(x, y, yerr=dy, color='silver', fmt='o-')
```

```python
plt.errorbar(x, y, yerr=dy, color='black', fmt='o',
            linewidth=1.9, ecolor='lightgray')
plt.tight_layout()
```

## Personnaliser l'affichage

```python
x = np.random.randn(100)

plt.style.use('classic')
fig = plt.figure(figsize=(5, 3))
ax = plt.axes(facecolor='#E6E6E6')

# Afficher les ticks en dessous de l'axe
ax.set_axisbelow(True)

# Cadre en blanc
plt.grid(color='w', linestyle='solid')

for spine in ax.spines.values():
    spine.set_visible(False)
    

# Cacher les marqueurs en haut et  droite
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# Personnaliser les tiquettes des marqueurs
# et leur appliquer une rotation
marqueurs = [-3, -2, -1, 0, 1, 2, 3]
xticklabels = list("ABCDEFG")
plt.xticks(marqueurs, xticklabels, rotation=-45)

for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')

# Changer les couleurs de bars
plt.hist(x, color="#EE6666", edgecolor="#E6E6E6")
ax.set_xlabel("Intervalles de valeurs", color='1')
#ax.set(xlabel='Intervalle de valeurs', ylabel='Amplitude de classes', color="1")
ax.set_ylabel("Amplitudecde classes", color='1')
plt.show()
```

```python
y = np.linspace(0, 10, 10)
plt.axes(facecolor='0.3')
plt.plot(y, np.sin(y),'-o', color='orange', linewidth=2, markersize=10, label='$\sin(x)$')
plt.plot(y, np.cos(y),'-s', label="$\cos(x)$")
plt.grid(color="silver")
plt.axis(False)
plt.legend(loc='center right')
plt.tight_layout()
```
