**Visualiser la fonction sinus**

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.xlim([0, 10])
plt.ylim([-1, 1.01])
```

```python
plt.plot(x, np.sin(x))
```

```python
fig = plt.figure()
ax = plt.axes()

plt.plot(x, np.sin(x), color='blue', linestyle='solid', label="blue")
plt.plot(x, np.sin(x - 1), color='g', linestyle='dashed', label='green')
plt.plot(x, np.sin(x - 2), color='0.7', linestyle='dashdot', label="gris")
plt.plot(x, np.sin(x - 3), color="#FF0000", linestyle='dotted', label="rouge")
plt.axis([0, 10, -1, 1])
plt.title("Example de graphes")
#plt.legend(loc="lower left")
plt.legend(loc=[1, 0], ncol=2)
ax.set(xlabel='x', ylabel='sin(x)')
plt.show()
```

## Visualiser l'incertitude
### Donnes Discrte

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy + np.random.randn(50)
```

```python
# plot the error
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='silver')
plt.tight_layout()
```

```python
plt.errorbar(x, y, yerr=dy, color='silver', fmt='o-')
```

```python
plt.errorbar(x, y, yerr=dy, color='black', fmt='o',
            linewidth=1.9, ecolor='lightgray')
plt.tight_layout()
```
