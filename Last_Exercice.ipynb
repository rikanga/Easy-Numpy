**Créer un décalage de 10 ans**

```python
# Créer un décade
births['decade'] = 10 * (births.year // 10)
```

**Convert day, month, year to be integer**

```python
births.dropna(inplace=True) # On supprime les valeurs nulles
```

```python
# Convertir le type des valeurs
for field in ['day', 'month', 'year']:
    births[field] = births[field].astype(int)
```
