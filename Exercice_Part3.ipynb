```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
hubble = pd.read_csv('hubble_data.csv')
```

```python
hubble
```

```python
hubble.plot(x='distance', y='recession_velocity')
```

```python
plt.plot(hubble.distance, hubble.recession_velocity)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
linear.fit(hubble.recession_velocity.values.reshape(-1, 1), hubble.distance.values.reshape(-1, 1))
```

```python
linear.coef_
```

```python
sns.regplot(hubble.distance, hubble.recession_velocity)
```

```python
x = hubble.distance.values.reshape(-1, 1)
y = hubble.recession_velocity
```

```python
linear = LinearRegression().fit(x, y)
```

```python
X_pred = linear.predict(x)
```

```python
sns.set()
plt.rcParams.update({'font.size':20})
plt.title("Donnes de Hubble")
plt.scatter(hubble.distance, hubble.recession_velocity, label="Donnes")
plt.plot(x, X_pred, color='red', label="Prdiction")
plt.legend(); plt.show()
```

```python
sns.set()
plt.rcParams.update({'font.size':20})
plt.title("Donnes de Hubble")
plt.scatter(hubble.distance, hubble.recession_velocity, label="Donnes")
plt.plot(x, X_pred, color='red', label="Prdiction")
plt.axis(False)
plt.legend(); plt.show()
```

```python
sns.regplot(x=hubble.distance, y=hubble.recession_velocity)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
plt.plot(hubble.distance, hubble.recession_velocity, 's-')
plt.plot(x, X_pred, color="red")
ax.set(xlabel='Velocit', ylabel="Distance", title="Donnes de Hubble")
plt.show()
```
