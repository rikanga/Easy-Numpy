```python
import numpy as np, matplotlib.pyplot as plt
#from math import pi, sqrt, exp,
import random
```

```python
def gaussian(x, mu, sigma):
    """
        Fonction qui fait le calcule de la densit√© gaussienne.
    """
    return 1/(sqrt(2 * pi * pow(sigma, 2))) * exp(-pow((x-mu), 2) / pow(2*sigma, 2))
```

 #### Example de test

```python
domaine = range(-100, 100)
```

```python
y = [gaussian(x, mu=0, sigma=20) for x in domaine]
```

```python
# Calcule du domaine
plt.plot(domaine, y)
```

### Echantillanage de donn√©es

```python
sample = np.random.rand(100, 200)
```

Nous avons:

* 100 echantillons
* 200 individus

**Somme sur des lignes**

```python
somme_lignes = np.sum(sample, 0)
somme_lignes
```

**Somme sur colonne**

```python
sommes_colonnes = np.sum(sample, 1)
sommes_colonnes
```

**Somme globale des √©l√®ments**

```python
np.sum(sample)
```

**Taille de la matrice**

```python
print("Taille de la matrice: {}".format(sample.shape))
print("Taille de la matrice apr√®s la somme sur ligne: {}".format(somme_lignes.shape))
print("Taille de la matrice apr√®s la somme sur les colonnes: {}".format(sommes_colonnes.shape))
```

## Simulation de Numpy avec le jeu de Monty Hall

```python
from enum import Enum
```

```python
class Strategie(Enum):
    CHANGER = 0
    GARDER = 1
```

```python
def play_game(strategie):
    
    # Toute les portes disponibles
    portes  = list(range(3))
    
    # Bonne porte
    bonne_porte = np.random.randint(0, 2)
    
    # Premier choix de la porte
    premier_choix = portes[np.random.randint(0, 1)]
    
    if premier_choix == bonne_porte:
        portes.remove(portes[np.random.randint(0, 1)])
    else:
        portes = [bonne_porte]
    
    # Deuxi√®me choix du joueur
    deuxieme_choix = 0
    if strategie == Strategie.GARDER:
        deuxieme_choix = portes[0]
    elif strategie == Strategie.CHANGER:
        deuxieme_choix = premier_choix
    else:
        raise ValueError("Cette valeur ne pas pris en charge")
        
```
