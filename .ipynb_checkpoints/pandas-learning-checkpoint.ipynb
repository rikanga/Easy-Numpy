**Fonction pour générer rapidement de dataframe**

```python
def make_df(col, ind):
    data = {c: [str(c) + str(i) for i in ind] for c in col}
    return pd.DataFrame(data)
```

```python
make_df('ABC', range(3))
```

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'department': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'date': [2004, 2008, 2012, 2014]})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="employee")
```

```python
df2.rename(columns={"employee":"emp_name"}, inplace=True)
df2
```

```python
pd.merge(df1,df2, left_on='employee', right_on='emp_name')
```
